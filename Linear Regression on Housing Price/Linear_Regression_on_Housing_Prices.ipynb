# Task: Linear Regression on Housing Prices
 Implement linear regression to predict housing prices based on features like
 square footage, number of bedrooms, and location.

```python
# Importing required libraries and packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Loading the dataset
housing = fetch_california_housing()
data = pd.DataFrame(housing.data, columns = housing.feature_names)
data['Price'] = housing.target
data.head()
```

```python
# Getting information of Dataset
data.info()
```

```python
# Splitting the data into training and testing sets
x = data.drop('Price', axis = 1)
y = data['Price']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# Training the model
train_model = LinearRegression()
train_model.fit(x_train, y_train)
```

```python
# Making predictions on test sets
y_predict = train_model.predict(x_test)
```

```python
# Evaluating the model's performance
mean_square = mean_squared_error(y_test, y_predict)
mean_absolute = mean_absolute_error(y_test, y_predict)
print(f"Mean Squared Error: {mean_square}")
print(f"Mean Absolute Error: {mean_absolute}")
```

```python
# Selecting relevant features by using correlation analysis
import seaborn as sns
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot = True)
plt.show()
```

```python
# Visualizing the results
plt.scatter(y_test, y_predict)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price v/s Predicted Price")
plt.show()
```
